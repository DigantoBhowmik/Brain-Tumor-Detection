```python
import matplotlib.pyplot as plt
import numpy as np

datasets = ['Scratched CNN', 'MobileNetV2', 'ResNet50', 'VGG19']
models = ['Random Forest', 'KNN', 'AdaBoost']
accuracies = np.array([
    [97.45, 87.84, 92.37, 94.74],
    [91.10, 89.47, 88.02, 92.01],
    [94.74, 75.31, 81.85, 86.38]
])

fig, ax = plt.subplots()
index = np.arange(len(datasets))
bar_width = 0.2
opacity = 0.8

for i in range(len(models)):
    ax.bar(index + i * bar_width, accuracies[i], bar_width,
           alpha=opacity, label=models[i])
    for j in range(len(datasets)):
        ax.text(index[j] + i * bar_width, accuracies[i][j] + 0.06, str(accuracies[i][j]), ha='center')

ax.set_xlabel('Orginal Dataset')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of CNN Models on Original Dataset')
ax.set_xticks(index + bar_width * 2)
ax.set_xticklabels(datasets)
ax.legend()
ax.set_ylim([60, 102])
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

datasets = ['Scratched CNN', 'MobileNetV2', 'ResNet50', 'VGG19']
models = ['Random Forest', 'KNN', 'AdaBoost']
accuracies = np.array([
    [98.91, 89.47, 91.47, 90.74],
    [90.56, 90.56, 84.39, 93.28],
    [96.55, 78.58, 81.12, 85.66]
])

fig, ax = plt.subplots()
index = np.arange(len(datasets))
bar_width = 0.2
opacity = 0.8

for i in range(len(models)):
    ax.bar(index + i * bar_width, accuracies[i], bar_width,
           alpha=opacity, label=models[i])
    for j in range(len(datasets)):
        ax.text(index[j] + i * bar_width, accuracies[i][j] + 0.06, str(accuracies[i][j]), ha='center')

ax.set_xlabel('Applying Laplacian Filter')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of CNN Models on Laplacian Filter')
ax.set_xticks(index + bar_width * 2)
ax.set_xticklabels(datasets)
ax.legend()
ax.set_ylim([60, 102])
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
from tabulate import tabulate

table = []
table.append([''] + models) # add column headers
for i in range(len(datasets)):
    row = [datasets[i]]
    for j in range(len(models)):
        row.append(accuracies[j][i])
    table.append(row)

print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))
```

```python
best_models = []
for i in range(len(datasets)):
    dataset_accuracies = accuracies[:, i]
    sorted_indices = np.argsort(dataset_accuracies)[::-1]  # sort in descending order
    best_models.append([models[sorted_indices[0]], models[sorted_indices[1]]])

table = [['Dataset', 'Best Model 1', 'Best Model 2']]
for i in range(len(datasets)):
    table.append([datasets[i], best_models[i][0], best_models[i][1]])
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))
```

```python
best_models = []
for i in range(len(datasets)):
    dataset_accuracies = accuracies[:, i]
    sorted_indices = np.argsort(dataset_accuracies)[::-1]  # sort in descending order
    best_models.append([models[sorted_indices[0]], models[sorted_indices[1]], 
                        models[sorted_indices[2]], models[sorted_indices[3]]])

table = [['Dataset', 'Best Model 1', 'Best Model 2', 'Best Model 3', 'Best Model 4']]
for i in range(len(datasets)):
    table.append([datasets[i], best_models[i][0], best_models[i][1], 
                  best_models[i][2], best_models[i][3]])
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))
```

```python
best_datasets = []
for i in range(len(models)):
    model_accuracies = accuracies[i]
    sorted_indices = np.argsort(model_accuracies)[::-1]  # sort in descending order
    best_datasets.append([datasets[sorted_indices[0]], datasets[sorted_indices[1]], 
                          datasets[sorted_indices[2]], datasets[sorted_indices[3]], 
                          datasets[sorted_indices[4]]])

table = [['Model', 'Best Dataset 1', 'Best Dataset 2', 'Best Dataset 3', 'Best Dataset 4', 'Best Dataset 5']]
for i in range(len(models)):
    table.append([models[i], best_datasets[i][0], best_datasets[i][1], 
                  best_datasets[i][2], best_datasets[i][3], best_datasets[i][4]])
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))
```

```python
best_models = []
best_accuracy = []
for i in range(len(datasets)):
    dataset_accuracies = accuracies[:, i]
    sorted_indices = np.argsort(dataset_accuracies)[::-1]
    best_accuracy.append(dataset_accuracies[sorted_indices[0]])
    best_models.append([models[sorted_indices[0]]])

table = [['Dataset', 'Model', 'Accuracy']]
for i in range(len(datasets)):
    table.append([datasets[i], best_models[i][0],best_accuracy[i]])
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))
```

```python
import matplotlib.pyplot as plt

# assume that the best_models and best_accuracy variables are already defined

fig, ax = plt.subplots()
dataset_labels = [f'{dataset}\n({accuracy:.2f}%)' for dataset, accuracy in zip(datasets, best_accuracy)]
x_pos = range(len(datasets))
accuracies = [accuracy.item() for accuracy in best_accuracy]
models = [model[0] for model in best_models]

ax.bar(x_pos, accuracies, align='center', alpha=0.5)
ax.set_xticks(x_pos)
ax.set_xticklabels(dataset_labels)
ax.set_ylabel('Accuracy')
ax.set_title('Best Model Accuracy by Dataset')

for i, v in enumerate(accuracies):
    ax.text(i, v + 0.5, f'{models[i]}', ha='center', rotation=90)
ax.set_ylim([95, 100])
plt.show()
```
